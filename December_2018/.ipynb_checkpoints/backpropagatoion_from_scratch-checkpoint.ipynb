```python
import numpy as np
```

```python
# data, predicions, weight, bias
x=np.array([1,2,3,4,5])
y=np.array([4,6,8,10,12])
w=0
b=0
```

```python
def prediction():
    y_=w*x+b
    return y_

def update_rule(lr,y_):
    global w,b
    dw=(2*(y_-y)*x)/5
    w-=lr*dw
    db=(2*(y_-y))/5
    b-=lr*db

def calculate_loss(y_):
    loss=(y_-y)**2
    return loss
    
def runner(epochs,lr):
    for epoch in range(epochs):
        pred=prediction()
        error=calculate_loss(pred)
        update_rule(lr=lr,y_=pred)
        print('\nNo of epoch={epoch},loss={loss},w={w},b={b},pred={pred}'.format(pred=pred,epoch=epoch,loss=error,w=w,b=b))
```

```python
if __name__ == '__main__':
    #default values
    #no of epochs=1000
    #learning rate=0.1
    runner(epochs=1000,lr=0.1)
```
